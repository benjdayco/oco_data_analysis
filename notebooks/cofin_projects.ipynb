import pandas as pd
import numpy as np

#read data file (f)
cofin_grants_loan_data_raw = pd.read_csv("../data_files/cofin loan grants with year-2-18-2016.csv")

#remove records with all blank columns (noise from lotus notes)
cofin_grants_loan_data = cofin_grants_loan_data_raw.dropna(subset = ['Approval Numbers CoFin','Proj No'] , how='all')

```python
cofin_grants_loan_data
```

```python
cofin_grant_loan = cofin_grants_loan_data[['Year', 'Partner', 'Country Mnemonic', 'Department', 'Division','Total CF']]
```

```python
cofin_group_by_year = cofin_grant_loan.groupby('Year')
```

```python
cofin_group_by_year
```

```python
cofin_group_by_year.ngroups
```

```python
cofin_group_by_year.groups
```

```python
# a helper function to print the contents of the groups
def print_groups (groupobject):
# loop over all groups, printing the group name
# and group details
    for name, group in groupobject:
        print (name)
        print (group)
```

```python
print_groups(cofin_group_by_year)
```
