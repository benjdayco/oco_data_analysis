```python
import pandas as pd
import numpy as np
```

```python
#read data files: raw dump that includes product approval number
#
cofin_grant_loans_all_raw = pd.read_csv("data_files/cofin_grant_loans_all_raw_6-20-16.csv")
cofin_grant_loans_all_raw
```

```python
#remove records with all blank columns (noise from lotus notes)
cofin_grant_loans_all = cofin_grant_loans_all_raw.dropna(subset = ['Loan/Grant Title'] , how='all')
```

```python
cofin_grant_loans_all
```

```python
#
#convert approval numbers to 4 characters with trailing 0's
#

#temporary dataframe 
temp_df = pd.DataFrame()

#iterate and extract approval numbers
for index,row in cofin_grant_loans_all.iterrows():

    #temporary store approval number of current row
    approval_number_orig = str(row['Approval Numbers'])

    #split approval number into array by / character
    approval_number_split =  approval_number_orig.split("/")
    
    #
    #convert each approval number to 4 characters with trailing zeros
    #
    join_numbers = [] #temporary array for converted number
    
    #itererate through each approval number in array
    for number in approval_number_split:
                
        join_numbers.append(number.zfill(4))
        clean_approval_numbers = '/'.join(join_numbers)       
        
        new_row = row #store new_row in temporary Series
        new_row['Extracted Approval Number'] = clean_approval_numbers #add number as new column
        
        print new_row
        
        #insert current approval number into new dataframe with row[Project No]
        temp_df = temp_df.append(new_row, ignore_index=True)
        new_df = temp_df

        

        
    
        

 
```
